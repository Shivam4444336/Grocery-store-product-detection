**This script counts number of products in each shelf image**

```python
# Download ShelfImages.tar.gz for test images
import requests
file_url = 'https://storage.googleapis.com/open_source_datasets/ShelfImages.tar.gz'
	
r = requests.get(file_url, stream = True)
#create empty file in current directory with below name
with open('/content/shelfimage.tar.gz', "wb") as file:
	for block in r.iter_content(chunk_size = 1024):
		if block:
			file.write(block)
```

```python
#Extract ShelfImages folder
import tarfile
tar_object = tarfile.open('/content/shelfimage.tar.gz')
#Extract ShelfImages folder in current directory
tar_object.extractall('/content')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
list_of_test_images_names = os.listdir('/content/drive/MyDrive/Final_Data/images/val')
```

```python
import cv2
import json
import torch
import numpy as np

image2products = {}
model = torch.hub.load('/content/yolov5', 'custom', path='/content/drive/MyDrive/yolo_v5_clearquote/runs/train/exp6/weights/best.pt', source='local') 
for index,image_name in enumerate(list_of_test_images_names):
  img = cv2.imread('/content/drive/MyDrive/Final_Data/images/val/'+image_name)
  results = model(img,size = 1280)
  dataframe = results.pandas().xyxy[0]
  number_of_products = np.sum(dataframe['class'] == 1)
  image2products[image_name+'_'+str(index)] = int(number_of_products)
```

```python
print(image2products)
```

```python
with open('/content/drive/MyDrive/image2products.json','w') as f:
  json.dump(image2products,f)
```
